```python
import seaborn as sns
from matplotlib import pyplot as plt

import fastf1
import fastf1.plotting

fastf1.plotting.setup_mpl(mpl_timedelta_support=False, misc_mpl_mods=False, color_scheme='fastf1')
```

2024 British Grand Prix Team Pace Comparison

```python
race = fastf1.get_session(2024, 'British Grand Prix', 'R')
race.load()
laps = race.laps.pick_quicklaps()

transformed_laps = laps.copy()
transformed_laps.loc[:, "LapTime (s)"] = laps["LapTime"].dt.total_seconds()

team_order = (
    transformed_laps[["Team", "LapTime (s)"]]
    .groupby("Team")
    .median()["LapTime (s)"]
    .sort_values()
    .index
)
print(team_order)

team_palette = {team: fastf1.plotting.get_team_color(team, session=race) for team in team_order}

fig, ax = plt.subplots(figsize=(15,10))
sns.boxplot(
    data=transformed_laps,
    x="Team",
    y="LapTime (s)",
    hue="Team",
    palette=team_palette,
    order=team_order,
    whiskerprops=dict(color="white"),
    boxprops=dict(edgecolor="white"),
    medianprops=dict(color="grey"),
    capprops=dict(color="white"),
)

plt.title("2024 British Grand Prix")
plt.grid(visible=False)

ax.set(xlabel=None)
plt.tight_layout()
plt.show()
```

Analysis:



*   Mclaren had fastest average laptime, around 1:31, with Mercedes close behind
*   Mercedes had smallest variability among the teams
- Ferrari had the quickest lap time overall, approximately in the low 1:28s
- Apline yielded the slowest median lap time, approximately at 93 seconds, almost a second and a half slower than Mclaren
